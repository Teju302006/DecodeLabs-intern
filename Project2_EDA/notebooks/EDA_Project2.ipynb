```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv(r"C:\Users\TejaswiniD\Downloads\archive (3)\sales.csv")

# Display First 5 Rows
print("First 5 Rows:")
print(df.head())

# Shape of Dataset
print("\nShape of Dataset:")
print(df.shape)

# Column Names
print("\nColumns:")
print(df.columns)

# Dataset Information
print("\nDataset Info:")
print(df.info())

# Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Statistical Summary
print("\nStatistical Summary:")
print(df.describe())

# Mean, Median, Count
print("\nSales Statistics")
print("Mean Sales:", df["Sales"].mean())
print("Median Sales:", df["Sales"].median())
print("Count Sales:", df["Sales"].count())

print("\nProfit Statistics")
print("Mean Profit:", df["Profit"].mean())
print("Median Profit:", df["Profit"].median())
print("Count Profit:", df["Profit"].count())
```

```python
market_sales = df.groupby("Market")["Sales"].sum()

print(market_sales)

market_sales.plot(kind="bar")
plt.title("Total Sales by Market")
plt.ylabel("Sales")
plt.show()
```

```python
product_sales = df.groupby("Product Type")["Sales"].sum()

print(product_sales)

product_sales.plot(kind="bar")
plt.title("Sales by Product Type")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Sales"])
plt.title("Sales Outliers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Profit"])
plt.title("Profit Outliers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Sales"].hist(bins=20)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Profit"].hist(bins=20)
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
