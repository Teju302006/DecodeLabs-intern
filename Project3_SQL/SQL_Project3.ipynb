```python
import pandas as pd
import sqlite3

df = pd.read_csv(r"C:\Users\TejaswiniD\Downloads\archive (3)\sales.csv")

conn = sqlite3.connect("sales.db")

df.to_sql("sales", conn, if_exists="replace", index=False)

print("Database Created Successfully")
```

```python
query = "SELECT * FROM sales LIMIT 10"

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT *
FROM sales
WHERE Profit > 100
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM sales
ORDER BY Sales DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM sales
ORDER BY Sales DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Total_Records
FROM sales
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Sales) AS Total_Sales
FROM sales
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(Profit) AS Average_Profit
FROM sales
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Market,
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Market
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Product Type],
       AVG(Profit) AS Avg_Profit
FROM sales
GROUP BY [Product Type]
"""

pd.read_sql(query, conn)
```
